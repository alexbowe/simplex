```python
import simplex as s
import itertools as it
import operator as op
import networkx as nx
```

```python
# absolute differences between tuple elements
def diffs(a, b):
    return map(abs, it.starmap(op.sub, it.izip(a,b)))

# there is an edge if the tuples differ by 1 in 1 dimension
is_edge = lambda a,b: reduce(op.add, diffs(a,b)) == 1

row = lambda i, dim: s.unique_perms(s.make_table(i, dim))
n_rows = lambda n, dim: [row(i, dim) for i in xrange(n)]

# Create dictionary of pascal's simplex edges (without the value)
def pascal_edges(n, dim):
    table = n_rows(n, dim)
    g = {}
    for row in table:
        for point, value in row:
            g[point] = []

    for point1 in g.keys():
        for point2 in g.keys():
            if is_edge(point1, point2):
                g[point1].append(point2)
                g[point2].append(point1)
    
    return g
```

```python
n, dim = 5, 3
G = nx.Graph(pascal_edges(n, dim))
nx.draw_graphviz(G, node_size=0)
```
